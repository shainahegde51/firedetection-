```python
def home_screen():
    homescreen = Toplevel(root)
    homescreen.title('fire-detection')
    width = 1350
    height = 770
    screen_width = homescreen.winfo_screenwidth()
    screen_height = homescreen.winfo_screenheight()
    x = (screen_width/2) - (width/2)
    y = (screen_height/2) - (height/2)
    homescreen.resizable(0, 0)
    homescreen.geometry("%dx%d+%d+%d" % (width, height, x, y))
    homescreen.configure(background = "white smoke")

    imgbg = PIL.Image.open("3.png")
    imgbg = imgbg.resize((1350, 770))
    photobg = ImageTk.PhotoImage(imgbg)
    lblhimgbg = Label(homescreen,image=photobg,bg="white smoke")
    lblhimgbg.image = photobg 
    lblhimgbg.place(x=0,y=0)

    lblhtitle=Label(homescreen,text="FIRE DETECTION",font=('times new roman', 40),fg='dark orange',bg='white smoke')
    lblhtitle.place(x=480,y=100)
    
    def fireinimage():
        imagescreen = Toplevel(homescreen)
        imagescreen.title('fire-detection')
        width = 1350
        height = 770
        screen_width = imagescreen.winfo_screenwidth()
        screen_height = imagescreen.winfo_screenheight()
        x = (screen_width/2) - (width/2)
        y = (screen_height/2) - (height/2)
        imagescreen.resizable(0, 0)
        imagescreen.geometry("%dx%d+%d+%d" % (width, height, x, y))
        imagescreen.configure(background = "white smoke")

        imgbg1 = PIL.Image.open("3.png")
        imgbg1 = imgbg1.resize((1350, 770))
        photobg1 = ImageTk.PhotoImage(imgbg1)
        lblhimgbg1 = Label(imagescreen,image=photobg1,bg="white smoke")
        lblhimgbg1.image = photobg1 
        lblhimgbg1.place(x=0,y=0)

        lblimgtitle=Label(imagescreen,text="FIRE DETECTION",font=('times new roman', 40),fg='dark orange',bg='white smoke')
        lblimgtitle.place(x=480,y=100)

        def openimage():
            filename = filedialog.askopenfilename(title = "Select A File", filetype =(("jpeg files","*.jpg"),("all files","*.*")),parent=imagescreen)
            imge = PIL.Image.open(filename)
            imge = imge.resize((420, 300))
            photoe = ImageTk.PhotoImage(imge)
            lblimge = Label(imagescreen,image=photoe)
            lblimge.image = photoe
            lblimge.place(x=480,y=290)

            def detectfireinimage():

                img = cv2.imread(filename)

                while True:
                    blur = cv2.GaussianBlur(img, (21, 21), 0)
                    hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)

                    lower = [18, 50, 50]
                    upper = [35, 255, 255]
                    lower = np.array(lower, dtype="uint8")
                    upper = np.array(upper, dtype="uint8")

                    mask = cv2.inRange(hsv, lower, upper)

                    output = cv2.bitwise_and(img, hsv, mask=mask)

                    no_red = cv2.countNonZero(mask)
                    
                    if int(no_red) > 15000:
                        Fire_Reported = Fire_Reported + 1
                    cv2.imwrite('output.png', output)
                    break
                cv2.destroyAllWindows()

                imgout = cv2.imread('output.png')
                lowcolor = (18, 50, 50)
                highcolor =(35, 255, 255)
                thresh = cv2.inRange(imgout, lowcolor, highcolor)
                count = np.sum(np.nonzero(thresh))
                #print(count)
                if count == 0:
                    messagebox.showinfo("Result", "NO FIRE DETECTED",parent=imagescreen)
                else:
                    messagebox.showinfo("Result", "FIRE DETECTED",parent=imagescreen)
                    
                os.remove('output.png')

            btnimagecheck=Button(imagescreen,text='Check',font=('times new roman', 20),height=1,fg='white smoke',bg='dark orange',command=detectfireinimage)
            btnimagecheck.place(x=645,y=600)

        btnimage=Button(imagescreen,text='Choose Image',font=('times new roman', 20),height=1,fg='white smoke',bg='dark orange',command=openimage)
        btnimage.place(x=600,y=230)

    def fireinvideo():
        videoscreen=Toplevel(homescreen)
        videoscreen.title('fire-detection')
        width = 1350
        height = 770
        screen_width = videoscreen.winfo_screenwidth()
        screen_height = videoscreen.winfo_screenheight()
        x = (screen_width/2) - (width/2)
        y = (screen_height/2) - (height/2)
        videoscreen.resizable(0, 0)
        videoscreen.geometry("%dx%d+%d+%d" % (width, height, x, y))
        videoscreen.configure(background = "white smoke")

        imgbg2 = PIL.Image.open("3.png")
        imgbg2 = imgbg2.resize((1350, 770))
        photobg2 = ImageTk.PhotoImage(imgbg2)
        lblhimgbg2 = Label(videoscreen,image=photobg2,bg="white smoke")
        lblhimgbg2.image = photobg2 
        lblhimgbg2.place(x=0,y=0)

        lblvidtitle=Label(videoscreen,text="FIRE DETECTION",font=('times new roman', 35),fg='dark orange',bg='white smoke')
        lblvidtitle.place(x=480,y=100)

        def openvideo():
            filename = filedialog.askopenfilename(title = "Select A File", filetype =(("all video format",".mp4"),("all files","*.*")),parent=videoscreen)
            filename=ntpath.basename(filename)
            lblvideopath1=Label(videoscreen,text="Your Selection:",font=('times new roman', 18),fg='dark orange',bg='white smoke')
            lblvideopath1.place(x=530,y=350)
            lblvideopath=Label(videoscreen,text="",font=('times new roman', 16),fg='dark orange',bg='white smoke')
            lblvideopath.place(x=700,y=350)

            if(filename==""):
                lblvideopath.config(text="No file chosen")

            else:
                lblvideopath.config(text=filename)

            def detectfireinvideo():
                video=cv2.VideoCapture(filename)

                while True:
                    (grabbed,frame)=video.read()
                    if not grabbed:
                        break
                    blur=cv2.GaussianBlur(frame,(21,21),0)
                    hsv=cv2.cvtColor(blur,cv2.COLOR_BGR2HSV)
                    lower=[18,50,50]
                    upper=[35,255,255]
                    lower=np.array(lower,dtype="uint8")
                    upper=np.array(upper,dtype="uint8")
                    mask=cv2.inRange(hsv,lower,upper)
                    output=cv2.bitwise_and(frame,hsv,mask=mask)
                    no_red=cv2.countNonZero(mask)
                    if(int(no_red>20000)):
                        lblvideoresult1.place(x=500,y=590)
                        lblvideoresult.config(text="FIRE DETECTED")
                        lblvideoresult.place(x=600,y=590)
                        cv2.putText(output,'Fire detected',(5, 90),cv2.FONT_HERSHEY_SIMPLEX,2,(0, 0, 200), 2)
                    else:
                        lblvideoresult1.place(x=480,y=590)
                        lblvideoresult.config(text="NO FIRE DETECTED")
                        lblvideoresult.place(x=580,y=590)
                        cv2.putText(output,'No Fire detected',(5, 90),cv2.FONT_HERSHEY_SIMPLEX,2,(0, 0, 200), 2)
                    cv2.imshow("output",output)
                    if cv2.waitKey(1) & 0xFF == ord('q'):
                        break
                cv2.destroyAllWindows()
                video.release()

            btnvideocheck=Button(videoscreen,text='Check',font=('times new roman', 20),height=1,fg='white smoke',bg='dark orange',command=detectfireinvideo)
            btnvideocheck.place(x=620,y=470)

        lblvideoresult1=Label(videoscreen,text="Result:",font=('times new roman', 22),fg='dark orange',bg='white smoke')

        lblvideoresult=Label(videoscreen,text="",font=('times new roman', 22),fg='dark orange',bg='white smoke')

        btnvideo=Button(videoscreen,text='Choose Video File',font=('times new roman', 20),height=1,fg='white smoke',bg='dark orange',command=openvideo)
        btnvideo.place(x=550,y=230)
        
    def fireinreal():
        video=cv2.VideoCapture(0)
        while True:
            (grabbed,frame)=video.read()
            if not grabbed:
                break
            blur=cv2.GaussianBlur(frame,(21,21),0)
            hsv=cv2.cvtColor(blur,cv2.COLOR_BGR2HSV)
            lower=[18,50,50]
            upper=[35,255,255]
            lower=np.array(lower,dtype="uint8")
            upper=np.array(upper,dtype="uint8")
            mask=cv2.inRange(hsv,lower,upper)
            output=cv2.bitwise_and(frame,hsv,mask=mask)
            no_red=cv2.countNonZero(mask)
            #print(no_red)
            if(int(no_red>20000)):
                lblrealresult1.place(x=550,y=520)
                lblrealresult.config(text="FIRE DETECTED")
                lblrealresult.place(x=640,y=520)
                cv2.putText(output,'Fire detected',(5, 90),cv2.FONT_HERSHEY_SIMPLEX,2,(0, 0, 200), 2)
            else:
                lblrealresult1.place(x=520,y=520)
                lblrealresult.config(text="NO FIRE DETECTED")
                lblrealresult.place(x=610,y=520)
                cv2.putText(output,'No Fire detected',(5, 90),cv2.FONT_HERSHEY_SIMPLEX,2,(0, 0, 200), 2)
            cv2.imshow("output",output)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        cv2.destroyAllWindows()
        video.release()
        
        
    btnimage=Button(homescreen,text='DETECT FIRE IN IMAGE',command=fireinimage,font=('times new roman', 25),height=1,fg='white smoke',bg='dark orange')
    btnimage.place(x=280,y=260)
    btnvideo=Button(homescreen,text='DETECT FIRE IN VIDEO',command=fireinvideo,font=('times new roman', 25),height=1,fg='white smoke',bg='dark orange')
    btnvideo.place(x=730,y=260)
    btnreal=Button(homescreen,text='REAL-TIME FIRE DETECTION',command=fireinreal,font=('times new roman', 25),height=1,fg='white smoke',bg='dark orange')
    btnreal.place(x=460,y=400)
    
    lblrealresult1=Label(homescreen,text="Result:",font=('times new roman', 22),fg='dark orange',bg='white smoke')
    lblrealresult=Label(homescreen,text="",font=('times new roman', 22),fg='dark orange',bg='white smoke')

```
